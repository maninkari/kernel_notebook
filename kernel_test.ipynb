```python
import math
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import HTML, display
from PIL import Image, ImageDraw
```

```python
x = np.linspace(start=0, stop=20, num=100)

def y(t):
    return math.exp(-t)*math.cos(math.pi*t)

yarr = []
for s in range(100):
    yarr.append(y(s/20))

plt.plot(x, yarr)
plt.show()
```

```python
L = 5

def kernel_f(n, m, l):
    h = math.floor(l/2)
    rad = math.sqrt((n-h)*(n-h) + (m-h)*(m-h))  
    return int(round(120*math.exp(-rad)*math.cos(rad*math.pi*3), 0))
    # return rad

tab = []
min = 0
max = 0
d = 0

for i in range(L):
    r = []
    for j in range(L):
        t = kernel_f(i, j, L)
        if min > t:
            min = t
        if max < t:
            max = t
        r.append(t)
    print(r)
    tab.append(r)

d = max - min
   
```

```python
sq_len = 20
im = Image.new('RGB', (L*sq_len, L*sq_len), (255, 255, 255))
draw = ImageDraw.Draw(im)
# draw.rectangle((0,0, sq_len, sq_len), fill=(0,0,0))

tab2 = [[0] * L for i in range(L)]

for i in range(L):
    for j in range(L):
        t = int(round(255*(tab[i][j]-min)/d))
        tab2[i][j] = t
        draw.rectangle((i*sq_len,j*sq_len, (i+1)*sq_len,(j+1)*sq_len), fill=(t,t,t))
im
```
